```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv('DecisionTreeDataset.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
df.columns
```

```python
y=df["Job Offered_Yes"]
x=df[['CGPA_Low', 'CGPA_Medium', 'Communication_Good', 'Apptitude_Low',
       'Programming Skill_Good']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=10,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
TN=cm[0][0]
FP=cm[0][1]
FN=cm[1][0]
TP=cm[1][1]
```

```python
accuracy=(TP+TN)/(TP+TN+FP+FN)
accuracy 
```

```python
ER=1-accuracy
ER
```

```python
SEN=TP/(TP+FN)
SEN
```

```python
SPE=(TN)/(TN+FP)
SPE
```

```python
# alternative way of accuray and sensitivity
from sklearn.metrics import accuracy_score,recall_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
# sensitivity
SEN=recall_score(y_test,y_pred)
SEN
```

```python
precision=TP/(TP+FP)
precision
```

```python
plot_tree(model)
plt.show()
```

```python
# classification report
print(classification_report(y_test,y_pred))
```
