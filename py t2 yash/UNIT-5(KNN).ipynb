```python
# TOTAL=TP+TN+FP+FN
# Model Accuracy
accuracy=(TP+TN)/TOTAL
```

```python
# ERROR RATE
ER=(FP+FN)/TOTAL
# OR
ER=1-accuracy
```

```python
# Sensitivity or recall score
SEN=(TP)/TP+FN
```

```python
# Specificity
SPE=(TN)/TN+FP
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
y=df["Outcome"]
x=df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
614**0.5   #to find number of n_neighbors 
```

```python
nn=KNeighborsClassifier(n_neighbors=25)
```

```python
model=nn.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
TN=cm[0][0]
TN
```

```python
FP=cm[0][1]
FP
```

```python
FN=cm[1][0]
FN
```

```python
TP=cm[1][1]
TP
```

```python
accuracy=(TP+TN)/(TP+TN+FP+FN)
accuracy 
```

```python
ER=1-accuracy
ER
```

```python
SEN=(TP)/TP+FN
SEN
```

```python
SPE=(TN)/(TN+FP)
SPE
```

```python
# PRE=4
# GLOCOSE=95
# BP=55
# SKINTHIKNESS=30
# INSULIN=180
# BMI=29
# DPF=0.5
# AGE=51
Prediction=model.predict([[4,95,55,30,180,29,0.5,51]])
Prediction
```
